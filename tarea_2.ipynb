5. Actualmente, las variables `dep_time` y `arr_time` son convenientes para observar, pero difíciles de usar para cálculos porque no son números realmente continuos. Conviértelos en el número de minutos desde la medianoche en nuevas variables llamadas `arr_time_min` y `dep_time_min`.

```python
# Solución
# 5. Convertir dep_time y arr_time en minutos desde la medianoche
def time_to_minutes(time):
if pd.isna(time):  # Manejo de valores nulos
        return None
    hours = int(time // 100)
    minutes = int(time % 100)
    return hours * 60 + minutes

flights_data['dep_time_min'] = flights_data['dep_time'].apply(time_to_minutes)
flights_data['arr_time_min'] = flights_data['arr_time'].apply(time_to_minutes)

# Verificar los nuevos valores
flights_data[['dep_time', 'dep_time_min', 'arr_time', 'arr_time_min']].head()
```

6. Esperamos que `air_time` y la diferencia `arr_time_min` - `dep_time_min` sean iguales, pero no lo son. ¿Por qué no? Identifica un problema importante con los datos y uno más sutil.

```python
# Solución
flights_data['time_difference'] = flights_data['arr_time_min'] - flights_data['dep_time_min']
flights_data['time_check'] = flights_data['time_difference'] - flights_data['air_time']

# Identificar problemas en los datos
major_issues = flights_data[flights_data['time_check'].notna() & (flights_data['time_check'] != 0)]

# Mostrar ejemplos y conteo de problemas
issue_count = major_issues.shape[0]
major_issues[['dep_time_min', 'arr_time_min', 'air_time', 'time_difference', 'time_check']].head(), issue_count
```

8. Calcula los retrasos promedio por destino para vuelos que se originan en Nueva York (NYC). Crea una variable que clasifique los destinos de peor a mejor, donde 1 representa el peor lugar en el ranking.







```python
# Solución
nyc_flights = flights_data[flights_data['origin'].isin(['JFK', 'LGA', 'EWR'])]

# Rellenar valores NaN en 'arr_delay'
nyc_flights['arr_delay'] = nyc_flights['arr_delay'].fillna(0)
dest_avg_delay = (nyc_flights.groupby('dest')['arr_delay'].mean().sort_values(ascending=False).reset_index())
dest_avg_delay['rank'] = dest_avg_delay['arr_delay'].rank(ascending=False, method='dense').astype(int)
print("Retrasos promedio por destino:")
print(dest_avg_delay.head())
```

9. Calcula la demora promedio por aerolínea

```python
# Solución
airline_avg_delay = (nyc_flights.groupby('carrier')['arr_delay'].mean().sort_values(ascending=False).reset_index())
print("Retraso promedio por aerolínea:")
print(airline_avg_delay.head())
```
